# Облака точек
Перед построением облака точек будет произведена очистка от стоп слов, в противном случае в визуализацию попадут только они)

## Облако точек для контекста

```python
df1 = pd.concat([df_train, df_test]).reset_index(drop = True).fillna("")
```

```python
import re 
all_text = ''.join(df1['context'].astype(str)).lower()

all_text = re.sub(r'[^\w\s]', '', all_text)

all_text = re.sub(r'\d+','', all_text)


stopWords = set(stopwords.words('english'))
words = word_tokenize(all_text)


wordsFiltered = [w for w in words if w not in stopWords]
words_count = Counter(wordsFiltered)
```

```python
most_popular = words_count.most_common(100)

text = " "
for word,n in tqdm(most_popular) :
    
    text+= " " + n*(word + " ")
    
```

```python
text[:100]
```

```python

x,y = [],[]
for a,b in most_popular[:20]:
    x.append(a)
    y.append(b)

plt.figure(figsize = (16,10))
plt.title("Top 20 самых встречаемых слов в context и их количество")
sns.barplot(x=y,y=x)
plt.show()
```

```python
Counter(text.split())
```

```python
wcloud = WordCloud(width = 800,collocations=False, height = 600, max_words = 100, background_color = "white").generate(text.lower())



plt.figure(figsize = (16,10))
plt.title("Облако точек для контекста")
plt.axis("off")
plt.imshow(wcloud)
plt.show()
```

# Визуализация эмбеддингов через T-SNE

## Визуализация эмбеддингов контекста

```python
embeddings = df['context'].progress_apply(lambda x: model.encode(x,device = 'cuda')).values
```

```python
# Применяем T-SNE для снижения размерности до 2D
tsne = TSNE(n_components=2, random_state=42)
embeddings_2d = tsne.fit_transform(np.array([i for i in embeddings]))


plt.figure(figsize=(16, 9))



plt.scatter(embeddings_2d[:,0],
            embeddings_2d[:,1])

plt.title('T-SNE Визулизация эмбеддингов')
plt.xlabel('T-SNE 1')
plt.ylabel('T-SNE 2')
plt.legend()
plt.grid(True)
plt.show()
```

# Проверка качества base-line retrieval'a

```python

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2').to("cuda")
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
```

```python
first_20_embed = model.encode(df['context'][:20])
```

```python
first_20_questions = model.encode(df['question'][:20])
```

# Косинусное расстояние для первых двадцати пар контекстов/вопросов.
   ### На диагонале размещены подходящие пары context/question.
   ### Как можно увидеть: base-line retriever дает довольно высокий скор даже просто на первых 256 токенах. Исключением является пара под индексом 8, несмотря на то, что релевантный контекст имеет наибольший скор,значение метрики получилось достаточно низким - это говорит о том, что в некоторых случая первых 256 токенов может быть недостаточно чтобы найти подходящие пары. 

```python
from sklearn.metrics.pairwise import cosine_similarity
plt.figure(figsize = (16,10))
plt.title("cosine_similarity")
sns.heatmap(cosine_similarity(first_20_embed, first_20_questions),annot = True)
```

# Сделаем такую же матрицу, но уже на все данные и посмотри статистике по главной диагонали .

```python
context_emb = model.encode(df['context'])
```

```python
question_emb = model.encode(df['question'])
```

```python
all_cosine_similaritys = cosine_similarity(context_emb,question_emb)
```

```python
import torch
torch.argmax(torch.randn(3,2), axis = 1)
```

```python
# найду идексы максимальный значений внутри векторов

argmax = np.argmax(all_cosine_similaritys, axis =1)
```

```python
# Поскольку релевантный контекст лежит на главной диагонали, подобным образом можно посчитать его accuracy, как можно увидить
# Почти в четверти случаев мы будем подтягивать не тот контекст в вопросу (Если брать top-1 чанк). 
sum(np.array([i for i in range(12000)]) == argmax)/12000
```
